```python
#importa framework Keras 
from keras.layers import Convolution2D, MaxPooling2D, Activation
from keras.models import Sequential
#importa numpy para os c√°lculos
import numpy as np
#importa matplotlib para "plotar" as imagens
import matplotlib.pyplot as plt
#importa cv2 para fazer o "read" da imagem
import cv2
```

```python
#Atrav√©s do m√≥dulo cv2 faz o "read" da imagem
cat = cv2.imread('E:\cat.png')
#"Plota" a imamgem na tela
plt.imshow(cat)
plt.show()
```

```python
#"Shape" da imagem
cat.shape
```

Camada de convolu√ß√£o com filtros 3X3 utilizando o modelo sequencial do Keras

```python
#Modelo sequ√™ncial Keras
model = Sequential()
# Camada de convolu√ß√£o com entrada cat.shape e filtros 3X3
model.add(Convolution2D(3,3,3,input_shape=cat.shape))
#Aumenta 1 dimens√£o
cat_batch = np.expand_dims(cat,axis=0)
#conv_cat recebe a sa√≠da das predi√ß√µes para o sample catch_batch
conv_cat = model.predict(cat_batch)
```

```python
#fun√ß√£o para plotar exibit o "shape" e "plotar" na tela a imagem da convolu√ß√£o
def visualize_cat(model, cat):
    cat_batch = np.expand_dims(cat,axis=0)
    conv_cat = model.predict(cat_batch)
    conv_cat = np.squeeze(conv_cat, axis=0)
    print(conv_cat.shape)
    plt.imshow(conv_cat)
    plt.show()
```

```python
#"Plota" imagem
visualize_cat(model, cat)
```

Camada de convolu√ß√£o com filtros 10X10 utilizando o modelo sequencial do Keras

```python
model = Sequential()
model.add(Convolution2D(3, 10,10,input_shape=cat.shape))
```

```python
#"Plota" imagem
visualize_cat(model, cat)
```

```python
#Fun√ß√£o para "Plotar" imagem ap√≥s Reshape
def imprime_gatinho_fofo(model, cat):
    cat_batch = np.expand_dims(cat,axis=0)
    conv_cat2 = model.predict(cat_batch)
    conv_cat2 = np.squeeze(conv_cat2, axis=0)
    print(conv_cat2.shape)
    conv_cat2 = conv_cat2.reshape(conv_cat2.shape[:2])
    print(conv_cat2.shape)
    plt.imshow(conv_cat2)
    plt.show()
```

Camada de convolu√ß√£o filtros 3X3N

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=cat.shape))
```

```python
#"Plotar" imagem 
imprime_gatinho_fofo(model, cat) 
```

Convolu√ß√£o com filtros 3X3 + camada ReLU

```python
model = Sequential()
model.add(Convolution2D(1,3,3,input_shape=cat.shape))
model.add(Activation('relu'))
```
